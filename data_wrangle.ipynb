```python
import pandas as pd

agri_data = pd.read_csv("data/agri_commodity_data.csv")
agri_data = agri_data[~agri_data["Commodity description"].str.contains("total", case=False)] # Remove totalled values from dataset
state_totaled_data = agri_data.groupby("Region label")["Gross value ($)"].sum()
```

```python
state_totaled_data = state_totaled_data.drop("Australia", axis=0)
```

```python
state_totaled_data.to_csv("data/state_sums.csv")
```

```python
state_totaled_data.sum() / 2
```

```python
state_by_category = agri_data.groupby(["Region label", "Category"])[["Gross value ($)"]].sum()
```

```python
state_by_category = state_by_category.drop("Australia", axis=0)
state_by_category = state_by_category.reset_index()
```
